```python
#importing required libraries
import requests
from bs4 import BeautifulSoup
```

```python
#requesting webpage server to get the source code
amazon=requests.get('https://www.amazon.in/s?i=electronics&bbn=1389432031&rh=n%3A1389432031%2Cp_36%3A-2000000&dc&qid=1612704765&rnid=1318502031&ref=sr_nr_p_36_1')
```

```python
#Postive 
amazon
```

```python
amazon.content
```

```python
#Parsing the source code using html.parser
soup=BeautifulSoup(amazon.content,'html.parser')
```

```python
soup
```

```python
#extracting tags of multiple pages
pages=soup.find_all('ul', class_='a-pagination')
pages
```

```python
#Creating a list of all the links of multiple pages
list=[]
for i in pages:
    links=soup.find_all('li')
    for j in range(1,len(links)):
        list.append(links[j])
list
```

```python
list1=[]
for k in list:
    a_tags=k.find_all('a')
    for p in a_tags:
        list1.append(p.get('href'))
list1
```

```python
#Fetching all the information of multiple pages at once . 
#Here fetching for a few pages only as the number of pages is quite large.
source=[]
for r in list1[1:6]:
    page_link=requests.get('https://www.amazon.in'+r)
    source.append(page_link.content)
source
```

```python
#Working on each page extracting the tags and scraping the information.
source[0]
source1=BeautifulSoup(source[0],'html.parser')

Name=[]
name=source1.find_all('span', class_='a-size-base-plus a-color-base a-text-normal')
for i in name:
    Name.append(i.text)

    
Price=[]
price=source1.find_all('span', class_='a-price-whole')
for j in price:
    Price.append(j.text)


    
Ratings=[]
rat=source1.find_all('span', class_='a-icon-alt')
for j in rat:
    Ratings.append(j.text)
Ratings=Ratings[5:]


Links=[]
lin=source1.find_all('a', class_='a-link-normal a-text-normal')
for i in lin:
        Links.append(i.get('href'))





print("lengths:")
print('Rating:',len(Ratings))
print('Price:',len(Price))
print('Name:' ,len(Name))
print('Links',len(Links))
```
