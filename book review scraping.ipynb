```python
import requests
from bs4 import BeautifulSoup
```

```python
#Requesting the webpage server to get the source code to scrape the data
page=requests.get("https://bookpage.com/reviews")
```

```python
page.content
```

```python
#Parsing the above source code using BeautifulSoup and parser - html.parser as the source code is in html
books=BeautifulSoup(page.content,'html.parser')
print(books.prettify())
```

```python
#Creatign empty lists which would store the records.
Book_Name=[]
Author=[]
Book_Genre=[]
Review=[]
```

```python
#extracting tags containing book name
name=books.find_all('h4',class_='italic')
name[0:5]
```

```python
#using for loop to get the records of the book name
for i in name:
    title=''
    for j in i.find_all('a'):
        title=title+j.get_text()
    print(title)
    Book_Name.append(title)
```

```python
#extracting tags containing author name
authors=books.find_all('p',class_="sans bold")
authors
```

```python
#fetching author names from the tags
for i in authors:
    Author.append(i.get_text().replace('\n',''))
```

```python
Author
```

```python
#Extracting tags containing book genre
genre=books.find_all('p',class_='genre-links hidden-phone')
genre
```

```python
#fetching book genre from the tags
for i in genre:
    Genre=''
    for j in i.find_all('a'):
        Genre=Genre+j.get_text()+", "
    print(Genre)
    Book_Genre.append(Genre)
```

```python
Book_Genre
```

```python
#Extracting review tags
Review=[]
list=['25835-syed-m-masood-bad-muslim-discount-fiction','25896-dayna-lorentz-feather-childrens','25912-annalee-newitz-four-lost-cities-history','25842-mark-harris-mike-nichols-biography','25921-martha-teichner-when-harry-met-minnie-biography-memoir']
for i in list:
    Book=requests.get('https://bookpage.com/reviews/'+i+'#.YBvIT-gzZPY')
    Books=BeautifulSoup(Book.content,'html.parser')
    review=Books.find('div',class_='article-body')
    print(review.get_text().replace('\n',''))
    Review.append(review.get_text().replace('\n',''))
```

```python
Review
```

```python
#creating a pandas dataframe to store the infromation scraped.
import pandas as pd
Book_Details=pd.DataFrame({})
Book_Details['Book_Name']=Book_Name[0:5]
Book_Details['Author']=Author[0:5]
Book_Details['Genre']=Book_Genre[0:5]
Book_Details['Review']=Review
```
